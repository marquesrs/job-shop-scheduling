```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
table = pd.read_csv("log.txt")

annealing = table[table["heuristica"] == "SimulatedAnnealing"]
localsearch = table[table["heuristica"] == "LocalSearchBest"]

print(f'Annealing\t time={annealing["tempo_ns"].mean()}\t makespan={annealing["makespan"].mean()}')
print(f'Local Search\t time={localsearch["tempo_ns"].mean()}\t makespan={localsearch["makespan"].mean()}')
```

```python
x_axis = [i for i in annealing["n"]]

y_axis = [t for t in localsearch["tempo_ns"]]

fig, ax = plt.subplots(dpi=200)

ax.plot(x_axis, y_axis, color="blue")

y_axis = [t for t in annealing["tempo_ns"]]

ax.plot(x_axis, y_axis, color="red")

plt.grid(True)
plt.show()
```
